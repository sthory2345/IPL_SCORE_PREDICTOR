```python
df["balls_bowled"] = (df["over"].astype(int) * 6) + df["ball"].astype(int)
df["balls_left"] = 120 - df["balls_bowled"]
df.head(126)
```

```python
df["player_dismissed"] = df["wicket"].astype(int)
df["player_dismissed"] = df.groupby(["match_no","innings"])["wicket"].cumsum()
df.tail(150)
```

```python
df["wicket_left"] = 10-df["player_dismissed"]
#df["wicket_left"] = df["wicket_left"].apply(lambda x: 0 if x < 0 else x)
df.head(150)
```

```python
df["current_run_rate"] = (df["current_score"] * 6) / df["balls_bowled"]
df.head(33)
```

```python
groups = df.groupby("match_no")
match_no = df["match_no"].unique()
last_five = []
for id in match_no:
    last_five.extend(groups.get_group(id)["runs"].rolling(window=30).sum().values.tolist())
df["last_five"] = last_five
df.head()
```

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from xgboost import XGBRegressor
from sklearn.metrics import r2_score, mean_absolute_error
import joblib
from google.colab import files

# Prepare your features and target
X = final_df.drop(columns=["runs_x"])
Y = final_df["runs_x"]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Define the transformer with handle_unknown=ignore in OneHotEncoder
transformer = ColumnTransformer([
    ("transformer", OneHotEncoder(drop="first", sparse_output=False, handle_unknown="ignore"), ["bat_team", "bowl_team", "venue"])
], remainder="passthrough")

# Definee the pipeline
pipe = Pipeline(steps=[
    ("step1", transformer),
    ("step2", StandardScaler()),
    ("step3", XGBRegressor(n_estimators=1000, learning_rate=0.2, max_depth=12, random_state=1))
])

# Train the pipeline
pipe.fit(X_train, Y_train)

# Pridict and evaluate the model
Y_pred = pipe.predict(X_test)
print("R2 Score:", r2_score(Y_test, Y_pred))
print("Mean Absolute Error:", mean_absolute_error(Y_test, Y_pred))

# Save the pipeline
joblib.dump(pipe, open("pipe.joblib", "wb"))

# Download the model
files.download('pipe.joblib')
```
